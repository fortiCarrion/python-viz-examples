```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set_theme() # adiciona tema padrao, para aplicar em todos os graficos
```

```python
df = pd.read_csv("data-visualization-graficos-bibliotecas-python\imigrantes_canada.csv")
```

```python
top_10 = df.sort_values(by="Total", ascending=False).head(10)
```

```python
sns.barplot(x="Total", y="PaÃ­s", data=top_10, orient="h")
```

```python
ax = sns.barplot(data=top_10, y="PaÃ­s", x="Total", orient="h", palette="vlag")
ax.set(title="Imigrantes para o Canada  \n 1980 a 2013", xlabel="Nr. de imigrantes", ylabel="")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_10, y="PaÃ­s", x="Total", orient="h")

ax.set_title("Paises com maior imigracao para o Canada\n1980 a 2013", loc="left", fontsize=16)
ax.set_xlabel("Nr. de imigrantes", fontsize=12)
ax.set_ylabel("")

plt.show()
```

```python
def gerar_grafico_paleta(palette, legend=True):
    fig, ax = plt.subplots(figsize=(8, 4))
    ax = sns.barplot(data=top_10, y="PaÃ­s", x="Total", orient="h", palette=palette, hue="PaÃ­s", legend=legend)

    ax.set_title("Paises com maior imigracao para o Canada\n1980 a 2013", loc="left", fontsize=16)
    ax.set_xlabel("Nr. de imigrantes", fontsize=12)
    ax.set_ylabel("")

    plt.show()
```

```python
gerar_grafico_paleta("Blues_r")
```

```python
gerar_grafico_paleta("rocket", legend=False)
```

```python
gerar_grafico_paleta("tab10", legend=False)
```

```python
sns.set_theme(style="dark")
gerar_grafico_paleta("tab10", legend=False)
```

```python
sns.set_theme(style="whitegrid")
gerar_grafico_paleta("tab10", legend=False)
```

```python
sns.set_theme(style="white")
gerar_grafico_paleta("tab10", legend=False)
```

```python
sns.set_theme(style="ticks")
gerar_grafico_paleta("tab10", legend=False)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_10, y="PaÃ­s", x="Total", orient="h", palette="tab10", hue="PaÃ­s")

ax.set_title("Paises com maior imigracao para o Canada\n1980 a 2013", loc="left", fontsize=16)
ax.set_xlabel("Nr. de imigrantes", fontsize=12)
ax.set_ylabel("")
sns.despine() # remove as caixas superiores..

plt.show()
```

```python
sns.set_theme()
sns.set_palette('Dark2')

fig, ax = plt.subplots(figsize=(10, 5))
ax = sns.lineplot(x="anos", y="imigrantes", data=_df.loc["Brasil"], label='Brasil', lw=3)
ax = sns.lineplot(x="anos", y="imigrantes", data=_df.loc["Argentina"], label='Argentina', lw=3)
ax = sns.lineplot(x="anos", y="imigrantes", data=_df.loc["Peru"], label='Peru', lw=3)
ax = sns.lineplot(x="anos", y="imigrantes", data=_df.loc["ColÃ´mbia"], label='ColÃ´mbia', lw=3)

ax.set_title('ImigraÃ§Ã£o dos maiores paÃ­ses da AmÃ©rica do Sul\npara o CanadÃ¡ de 1980 a 2013', loc='left', fontsize=20)
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('NÃºmero de imigrantes', fontsize=14)

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

ax.legend(title='PaÃ­ses', loc='upper right', bbox_to_anchor=(1.18, 1.02))

plt.show()
```
