```python
import pandas as pd
```

```python
df = pd.read_csv("data-visualization-graficos-bibliotecas-python\imigrantes_canada.csv")
```

```python
df.head()
```

```python
colunas = list(df.loc[:, "1980":"2013"].columns)

df.melt(id_vars=["Continente", "Região"], value_vars=colunas, value_name="emigrantes", var_name="ano")
```

```python
df.info()
```

```python
df.Continente.value_counts(normalize=True)
```

```python
df.set_index("País", inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc["Brasil", anos].copy()
```

```python
brasil
```

```python
brasil_dict = {"ano": brasil.index.tolist(),
               "imigrantes": brasil.values.tolist()}
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
df.head(2)
```

```python
import matplotlib.pyplot as plt
```

```python
dados_brasil.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(dados_brasil["ano"], dados_brasil["imigrantes"])
plt.title("Imigração do Brasil para Canada")
plt.xticks(["1980", "1985", "1990", "1995", "2000", "2005", "2010"])
plt.xlabel("ano")
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.ylabel("Qtd. de imigrantes")
plt.show()
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(13, 4))
ax.barh(america_sul.index, america_sul.Total, color=cores)
ax.set_title("Imigracao da America do Sul para o Canada \n 1980 a 2013", loc="left", fontsize=18)
ax.set_xlabel("Nr. de imigrantes", fontsize=13)
ax.set_ylabel("")
ax.xaxis.set_tick_params(labelsize=11)
```

```python
america_sul_sorted = america_sul.sort_values("Total", ascending=True)
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12, 4))
ax.barh(america_sul_sorted.index, america_sul_sorted.Total, color=cores)
ax.set_title("Imigracao da America do Sul para o Canada \n 1980 a 2013", loc="left", fontsize=18)
ax.set_xlabel("Nr. de imigrantes", fontsize=13)
ax.set_ylabel("")
ax.xaxis.set_tick_params(labelsize=11)
```

```python
cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else: cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 4))
ax.barh(america_sul_sorted.index, america_sul_sorted.Total, color=cores)
ax.set_title("Imigracao da America do Sul para o Canada \n 1980 a 2013", loc="left", fontsize=18)
ax.set_xlabel("Nr. de imigrantes", fontsize=13)
ax.set_ylabel("")
ax.xaxis.set_tick_params(labelsize=11)

for i, v in enumerate(america_sul_sorted["Total"]):
  ax.text(v + 20, i, str(v), color="black", fontsize=9, ha="left", va="center")

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis="both", which="both", length=0)

plt.plot()
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# Criando DataFrame
df_vendas_2022 = pd.DataFrame(vendas_2022, index=lojas)
```

```python
df_vendas_2022
```

```python
df_vendas_2022.index[0]
```

```python
fig, axs = plt.subplots(2,2, figsize=(10, 5))

# ajustar os espacamentos entre os subplots
plt.subplots_adjust(wspace=0.3, hspace=0.5)

# Adicionando um título geral para os subplots
fig.suptitle("Vendas por loja A,B,C,D \n periodo de 2022")

# Lista de cores
cores = ['darkviolet', 'green', 'darkblue', 'coral']

for i, ax in enumerate(axs.flat):
    #print(df_vendas_2022.loc[df_vendas_2022.index[i]])
    ax.plot(df_vendas_2022.loc[df_vendas_2022.index[i]], color=cores[i], lw=3)
    ax.set_title(f"Vendas da Loja: {df_vendas_2022.index[i]}")
    ax.set_xlabel("Mês", fontsize=10)
    ax.set_ylabel("Nr. de Vendas R$", fontsize=10)
    ax.tick_params(labelsize=8)
    ax.grid(color='lightgrey')

plt.show()
```
